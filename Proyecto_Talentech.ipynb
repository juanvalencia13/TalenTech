La Tasa Representativa del Mercado (TRM) y las tasas de interés de los bancos son indicadores fundamentales para comprender la economía de un país. La TRM refleja el valor del dólar estadounidense frente a la moneda local, y su estudio es crucial para empresas, inversionistas y consumidores que realizan transacciones internacionales o tienen deudas en moneda extranjera. Las fluctuaciones de la TRM afectan directamente los costos de importación y exportación, la competitividad de los productos nacionales y la balanza comercial, lo que a su vez influye en el crecimiento económico y la estabilidad financiera de una nación. Entender estos movimientos permite a los actores económicos tomar decisiones informadas y gestionar el riesgo cambiario de manera efectiva.

Por otro lado, las tasas de interés de los bancos son una herramienta clave de la política monetaria utilizada por los bancos centrales para controlar la inflación y estimular o enfriar la economía. Las tasas de interés afectan el costo del crédito y, por ende, el consumo y la inversión. Un aumento en las tasas de interés tiende a reducir el endeudamiento y el gasto, mientras que una disminución puede incentivar el consumo y la inversión, promoviendo el crecimiento económico. Estudiar estas tasas es esencial para los consumidores y empresas que buscan financiamiento, ya que les permite anticipar los costos futuros de sus préstamos y evaluar las oportunidades de inversión. Además, los analistas financieros y economistas utilizan esta información para prever el comportamiento de los mercados financieros y la economía en general.

```python
comp_fecha = df_union.groupby('Year_trm')['TRM'].mean()
plt.figure(figsize=(8, 4))
comp_fecha.plot(kind='bar', color='green')
plt.title('TRM MAXIMA POR AÑO')
plt.xlabel('Fecha')
plt.ylabel('TRM')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
# Iterate over the index and values of the comp_fecha Series instead of the trm DataFrame
for i, value in enumerate(comp_fecha.values):
    plt.text(i, value , f"{value:.0f}", ha='center') # Format the value as a floating-point number
plt.show()
```

```python
comp_fecha = df_union.groupby('year_interes')['valor_tasa'].mean()
plt.figure(figsize=(8, 4))
comp_fecha.plot(kind='bar', color='green')
plt.title('TRM MAXIMA POR AÑO')
plt.xlabel('Fecha')
plt.ylabel('TRM')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
# Iterate over the index and values of the comp_fecha Series instead of the trm DataFrame
for i, value in enumerate(comp_fecha.values):
    plt.text(i, value , f"{value:.2f}", ha='center') # Format the value as a floating-point number
plt.show()
```
